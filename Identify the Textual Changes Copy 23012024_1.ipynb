```python
#  Using difflib with HTML Output

import difflib

def generate_html_diff(text1, text2, output_file):
    
    # Split the texts into lines
    lines1 = text1.split('.')
    lines2 = text2.split('.')

    # Create an HtmlDiff object
    html_diff = difflib.HtmlDiff()

    # Generate the HTML diff
    html_content = html_diff.make_file(lines1, lines2, context=True, numlines=3)
    
    # Write the HTML content to a file
    with open(output_file, 'w', encoding='utf-8') as file:
        file.write(html_content)


generate_html_diff(text_edition1, text_edition2, 'diff_chapters.html')

# Display html file to see the visualisation
from IPython.display import HTML

# Specify the path to your HTML file
file_path = 'diff_chapters.html'

# Open and read the HTML file
with open(file_path, 'r', encoding='utf-8') as file:
    html_content = file.read()

# Display the HTML content in the notebook
HTML(html_content)
```
